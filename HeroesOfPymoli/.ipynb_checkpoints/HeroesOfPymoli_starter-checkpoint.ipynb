```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
total_players = purchase_data['SN'].nunique()
print('Total players: {}'.format(int(total_players)))
total_players
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Obtain unique items
no_unique_item = purchase_data['Item ID'].nunique()
#avg price
avg_price = round(purchase_data['Price'].mean(),2)
#obtain number of purchases
no_purchases = purchase_data['Item Name'].count()
#obtain total revenue
total_revenue = purchase_data['Price'].sum()

#Create summary datafrmae
summary={'Number of Unique Items':[no_unique_item],
        'Average Price':[avg_price],
        'Number of Purchases': [no_purchases],
        'Total Revenue': total_revenue}
data_summary = pd.DataFrame.from_dict(summary)

#Formatting the values
data_summary['Average Price'] = data_summary['Average Price'].map("${:,.2f}".format)
data_summary['Total Revenue'] = data_summary['Total Revenue'].map("${:,.2f}".format)
data_summary
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#Create a dataframe to hold all unique players by their name and remove the duplicates

unique = purchase_data.drop_duplicates(subset='SN')
unique = unique[['Gender']]

#Count no of players for each category
male = len([i for i in unique['Gender'] if i == 'Male'])
female = len([i for i in unique['Gender'] if i == 'Female'])
other = len([i for i in unique['Gender'] if i == 'Other / Non-Disclosed'])
total_count=[male, female, other]
# Create a dataframe with Gender Counts 
gender_demo = pd.DataFrame(np.array([['Male', male], ['Female', female], ['Other', other]]),
                    columns=['Gender', 'Total Count'])
gender_demo['Total Count']=pd.to_numeric(gender_demo['Total Count'])

# Create a new column in the gender_demo dataframe for calculating percentages
gender_demo['Percentage'] = ((gender_demo['Total Count']/gender_demo['Total Count'].sum())*100)
gender_demo["Percentage"] = gender_demo["Percentage"].map("{:.2f}%".format)
gender_demo
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Creating a dataframe for purchasing analysis
purchase_data_gender = pd.DataFrame(purchase_data.groupby("Gender")['Purchase ID'].count())
purchase_data_gender = purchase_data_gender.rename(columns={"Purchase ID":"Purchase Count"})

#Calculate other fields
purchase_data_gender['Average Purchase Price'] = pd.DataFrame(purchase_data.groupby("Gender")['Price'].mean())
purchase_data_gender['Total Purchase Value'] = pd.DataFrame(purchase_data.groupby("Gender")['Price'].sum())
purchase_data_gender['Average Total Purchase Per Person'] = purchase_data_gender['Total Purchase Value'] /purchase_data_gender['Purchase Count'] 

#format values
purchase_data_gender['Average Purchase Price'] = purchase_data_gender['Average Purchase Price'].map("${:,.2f}".format)
purchase_data_gender['Total Purchase Value'] = purchase_data_gender['Total Purchase Value'].map("${:,.2f}".format)
purchase_data_gender['Average Total Purchase Per Person'] = purchase_data_gender['Average Total Purchase Per Person'].map("${:,.2f}".format)

purchase_data_gender
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
df_age_bin = purchase_data
# Bins of ages   
bins = [0, 9, 14, 19, 24, 29,34,39,100]

# Create the names for the bins
group_names = ["<10", "10-14", "15-19", "20-24", "25-29","30-34","35-39","40+"]

#Create bins
df_age_bin["Age Ranges"] = pd.cut(df_age_bin["Age"], bins, labels=group_names)
df_age_bin = pd.DataFrame(df_age_bin.groupby('Age Ranges')['Item ID'].count())

#Rename the column to show Total Count value and convert it to numeric
df_age_bin = df_age_bin.rename(columns={"Item ID":"Total Count"})
df_age_bin['Total Count']= pd.to_numeric(df_age_bin['Total Count'])

#Calculate percentage of players
df_age_bin['Percentage of Players'] = (df_age_bin['Total Count']/df_age_bin['Total Count'].sum())*100

#Formatting the values
df_age_bin['Percentage of Players'] = df_age_bin['Percentage of Players'].map("{:,.2f}%".format)

df_age_bin
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
df_bin = purchase_data
# Bins of ages   
bins = [0, 9, 14, 19, 24, 29,34,39,100]

# Create the names for the bins
group_names = ["<10", "10-14", "15-19", "20-24", "25-29","30-34","35-39","40+"]

#Create bins
df_bin["Age Ranges"] = pd.cut(df_bin["Age"], bins, labels=group_names)

#Create new dataframe with group by Age ranges and Purchase Count for each range
df_ar = pd.DataFrame(df_bin.groupby('Age Ranges')['Item ID'].count())
df_ar= df_ar.rename(columns={"Item ID":"Purchase Count"})

#Create other columns
df_ar['Average Purchase Price'] = df_bin.groupby('Age Ranges')['Price'].mean()
df_ar['Total Purchase Value'] = df_bin.groupby('Age Ranges')['Price'].sum()
df_ar['Avg Total Purchase Per Person'] = df_ar['Total Purchase Value'] /df_ar['Purchase Count']

#Format output type into currency
df_ar['Average Purchase Price'] = df_ar['Average Purchase Price'].map("${:,.2f}".format)
df_ar['Total Purchase Value'] = df_ar['Total Purchase Value'].map("${:,.2f}".format)
df_ar['Avg Total Purchase Per Person'] = df_ar['Avg Total Purchase Per Person'].map("${:,.2f}".format)

df_ar
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Create new dataframe -group by the Name of Buyer and get count of diff items they purchased
top_df= pd.DataFrame(purchase_data.groupby('SN')['Item ID'].count())
top_df= top_df.rename(columns={"Item ID":"Purchase Count"})

#Calculate other variables
top_df['Average Purchase Price']= pd.DataFrame(purchase_data.groupby('SN')['Price'].mean())
top_df['Total Purchase Value']= pd.DataFrame(purchase_data.groupby('SN')['Price'].sum())

#sort dataframe in descending order based on their total purchase price
top_df = top_df.sort_values(by=['Total Purchase Value'],ascending = False)

#get topmost 5 buyers
top_df = top_df.head(5)


#Formatting the values
top_df['Average Purchase Price'] = top_df['Average Purchase Price'].map("${:,.2f}".format)
top_df['Total Purchase Value'] = top_df['Total Purchase Value'].map("${:,.2f}".format)

top_df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Create new dataframe -group by Item ID, Name, Price and count by Item ID
pop_df= pd.DataFrame(purchase_data.groupby(['Item ID','Item Name','Price'])['Item ID'].count())

#Rename Col to Purchase Count
pop_df= pop_df.rename(columns={"Item ID":"Purchase Count"})

#Count total Purchase value
pop_df['Total Purchase Value'] = pd.DataFrame(purchase_data.groupby(['Item ID','Item Name','Price'])['Price'].sum())

```

```python
#sort rows by purchase count and get first 5 top rows
pop_purchase_df = pop_df.sort_values(by=['Purchase Count'],ascending = False)
pop_purchase_df = pop_purchase_df.head(5)

#Formatting the values
#pop_df['Price'] = pop_df['Price'].map("${:,.2f}".format)
pop_purchase_df['Total Purchase Value'] = pop_purchase_df['Total Purchase Value'].map("${:,.2f}".format)

pop_purchase_df
```
